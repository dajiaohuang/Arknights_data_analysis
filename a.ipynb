```python
import matplotlib as mpl
import numpy as np

import pandas as pd
import matplotlib.pyplot as plt

import re  

import plotly.offline as pyo 
import plotly.graph_objects as go
import plotly.express as px
  
def extract_numbers(input_string):   
    pattern = re.search(r'\d+\.\d+', input_string)
    numbers = -100
    if pattern:
        numbers = pattern.group()
    return float(numbers)


df = pd.read_json('ArkOperators_final.json')



# print(df.head(1))
# print(df.生命上限.describe())

shuju = ['代号','身高','职业','生命上限','攻击','防御','法术抗性','再部署','部署费用','阻挡数','攻击速度']


a_values = df['生命上限'].str.split('/').str[0]  
b_values = df['生命上限'].str.split('/').str[1]  

  

df['初始生命值'] = list(map(int, a_values)) 
df['满级生命值'] = list(map(int, b_values)) 

a_values = df['攻击'].str.split('/').str[0]  
b_values = df['攻击'].str.split('/').str[1]  
  
df['初始攻击'] = list(map(int, a_values))
df['满级攻击'] = list(map(int, b_values)) 

a_values = df['防御'].str.split('/').str[0]  
b_values = df['防御'].str.split('/').str[1]  
  
df['初始防御'] = list(map(int, a_values))  
df['满级防御'] = list(map(int, b_values)) 

a_values = []
b_values = []

for i in df['法术抗性']:
    if '/' in i:
        index = i.find('/')
        a_values.append(i[0:index])
        b_values.append(i[index+1:])
    else:
        a_values.append(i)
        b_values.append(i)

  
df['初始法术抗性'] = a_values
df['满级法术抗性'] = b_values 

df.fillna(0, inplace=True)   

a_values = []
for i in df['再部署']:
    x = extract_numbers(i)
    a_values.append(x)


df['再部署时间'] = list(map(int, a_values))  

 


a_values = []

for i in df['部署费用']:
    if type(i) is int:
        a_values.append(19)
        continue
    if '/' in i:
        index = i.find('/')
        a_values.append(i[0:index])
    else:
        a_values.append(i)

df.fillna(0, inplace=True)  
df['部署费用'] = a_values


```

```python
# df.groupby('职业').满级生命值.mean()

df.groupby('职业').满级生命值.mean().plot(kind = "bar", color = ["blue"], figsize = [16,8], subplots = True)
plt.xticks(fontproperties='SimSun', rotation='horizontal', fontsize = 20)
plt.xlabel('职业', fontproperties='SimSun', fontsize = 20)
plt.yticks(fontsize = 16)
plt.title('分职业满级生命平均值', fontproperties='SimHei', fontsize = 30)
```
