```python
import pandas as pd
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt

df = pd.read_csv('ArkModified2.csv')
XF = []
FZ = []
JW = []
JJ = []
SS = []
TZ = []
YL = []
ZZ = []

for i in range(294):
    if df['职业'][i] == '先锋': XF.append(i)
    elif df['职业'][i] == '辅助': FZ.append(i)
    elif df['职业'][i] == '近卫': JW.append(i)
    elif df['职业'][i] == '狙击': JJ.append(i)
    elif df['职业'][i] == '术师': SS.append(i)
    elif df['职业'][i] == '特种': TZ.append(i)
    elif df['职业'][i] == '医疗': YL.append(i)
    elif df['职业'][i] == '重装': ZZ.append(i)

print(YL)
print(ZZ)
xf = df.loc[XF,].copy()
fz = df.loc[FZ,].copy()
jw = df.loc[JW,].copy()
jj = df.loc[JJ,].copy()
ss = df.loc[SS,].copy()
tz = df.loc[TZ,].copy()
yl = df.loc[YL,].copy()
zz = df.loc[ZZ,].copy()


df = df.loc[:, ['hp1','hp2','attack1','attack2','defence1','defence2']].copy()

print(df)
xf = xf.loc[:, ['hp1','hp2','attack1','attack2','defence1','defence2']].copy()
fz = fz.loc[:, ['hp1','hp2','attack1','attack2','defence1','defence2']].copy()
jw = jw.loc[:, ['hp1','hp2','attack1','attack2','defence1','defence2']].copy()
jj = jj.loc[:, ['hp1','hp2','attack1','attack2','defence1','defence2']].copy()   
ss = ss.loc[:, ['hp1','hp2','attack1','attack2','defence1','defence2']].copy()
tz = tz.loc[:, ['hp1','hp2','attack1','attack2','defence1','defence2']].copy()
yl = yl.loc[:, ['hp1','hp2','attack1','attack2','defence1','defence2']].copy()
zz = zz.loc[:, ['hp1','hp2','attack1','attack2','defence1','defence2']].copy()
print(zz)
```

```python

df.rename(columns={"hp1": "HP-Lv.0"}, inplace=True)
df.rename(columns={"hp2": "HP-Lv.max"}, inplace=True)
df.rename(columns={"attack1": "Attack-Lv.0"}, inplace=True)
df.rename(columns={"attack2": "Attack-Lv.max"}, inplace=True)
df.rename(columns={"defence1": "Defence-Lv.0"}, inplace=True)
df.rename(columns={"defence2": "Defence-Lv.max"}, inplace=True)
df.hist(alpha=0.5,bins = 100)
plt.tight_layout()
plt.show
```

```python
hp1mean,hp2mean,attack1mean,attack2mean,defence1mean,defence2mean = df['hp1'].mean(),df['hp2'].mean(),df['attack1'].mean(),df['attack2'].mean(),df['defence1'].mean(),df['defence2'].mean()
```

```python
hp1std,hp2std,attack1std,attack2std,defence1std,defence2std = df['hp1'].std(),df['hp2'].std(),df['attack1'].std(),df['attack2'].std(),df['defence1'].std(),df['defence2'].std()
```

```python
print('hp1正态性检验：\n',stats.kstest(df['hp1'], 'norm', (hp1mean, hp1std)))
print('hp2正态性检验：\n',stats.kstest(df['hp2'], 'norm', (hp2mean, hp2std)))
print('attack1正态性检验：\n',stats.kstest(df['attack1'], 'norm', (attack1mean, attack1std)))
print('attack2正态性检验：\n',stats.kstest(df['attack2'], 'norm', (attack2mean, attack2std)))
print('defence1正态性检验：\n',stats.kstest(df['defence1'], 'norm', (defence1mean, defence1std)))
print('defence2正态性检验：\n',stats.kstest(df['defence2'], 'norm', (defence2mean, defence2std)))
```

```python
import numpy as np
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
colo = ['b', 'g', 'r', 'c', 'm', 'y', 'k']
x = data.to_numpy()
x=x.transpose()
x=np.array([x[0],x[2],x[4]])
x=x.transpose()

fig = plt.figure(figsize=(12, 8))
ax = Axes3D(fig,  elev=30, azim=20)


shape = x.shape
sse = []
score = []
K = 4 # 分为K类
for k in [K]:
    clf = KMeans(n_clusters=k)
    clf.fit(x)
    sse.append(clf.inertia_)
    lab = clf.fit_predict(x)
    score.append(silhouette_score(x, clf.labels_, metric='euclidean'))
    for i in range(shape[0]):
        plt.xlabel('x')
        plt.ylabel('y')
        plt.title('k=' + str(k))
        ax.scatter(x[i, 0],x[i, 1], x[i, -1], c=colo[lab[i]])
    plt.show()
```
